## **Monitor, Protect and Preserve Wetland Ecosystems**

### **Data we are working with :**

| **Variable** | **Type** | **Source** |
|---------------|-----------|-------------|
| **Air Temperature (°C)** | Numerical | Onboard Thermometer |
| **Humidity (%)** | Numerical | AIR Sensor |
| **Infrared Radiation Index** | Numerical | IR Sensor |
| **Air Quality Index (AQI)** | Numerical | AIR Sensor / [OpenAQ API](https://openaq.org/#/api) |
| **Light Intensity (lux)** | Numerical | Camera Light Sensor |
| **Water Temperature (°C)** | Numerical | Thermometer (submersible probe) |
| **pH Level** | Numerical | External pH Sensor / [FAO Water Data](https://www.fao.org/aquastat/en/databases/water-use/) |
| **Dissolved Oxygen (mg/L)** | Numerical | Water Dataset API |
| **Turbidity (NTU)** | Numerical | Water Quality Dataset |
| **Nitrate & Phosphate (mg/L)** | Numerical | Open Environmental Datasets |
| **Water Surface Temperature Gradient** | Numerical | IR Camera Fusion |
| **Animal Movement** | Visual | Onboard Camera + Object Detection Model |
| **Species Classification** | Categorical | Camera + ML Model / [GBIF](https://www.gbif.org/) |
| **Anomaly Events (Disturbances)** | Boolean / Categorical | ML Trigger + Sensor Fusion |
| **GPS Coordinates** | Numerical | GPS Module |
| **Weather Context** | Categorical | [Open-Meteo API](https://open-meteo.com/) |
| **Season / Time of Day** | Categorical | Derived Variable |
| **Wildlife Activity Index** | Computed | Derived (Camera + Mic + IR) |
